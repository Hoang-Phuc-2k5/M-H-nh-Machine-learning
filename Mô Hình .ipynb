```python
# vẽ đồ thị sử dụng 2 thuộc tính còn lại mà không dùng target để phân loại màu 
plt.figure(figsize=(8, 6))
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], color='red', label='Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter plot of Sepal Length vs Sepal Width')
plt.legend()
plt.grid(True)
plt.show()
```

### Dựa vào các attribute ở trên để xây dựng cây


```python
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
plt.figure(figsize=(20,10))
tree.plot_tree(clf, 
               filled=True,          # Tô màu các nút theo lớp 
               feature_names=iris.feature_names,  # Sử dụng tên của các đặc trưng để làm nhãn cho các nút
               class_names=iris.target_names,     # Sử dụng tên của các lớp (target) làm nhãn cho các nút lá
               rounded=True,         # Tạo các hộp hình vuông
               proportion=True) # kích thước hộp tỉ lệ với số mẫu
plt.show()
```
